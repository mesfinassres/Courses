```python
# Start writing code here...
import numpy as np
from matplotlib import pyplot as plt
y= 200 + np.random.randn(200)
y
```

```python
y.shape
```

```python
x=[x for x in range(len(y))]
x
```

```python
plt.plot(x,y,'-')
```

```python
plt.fill_between(x,y,195, where=(y>195), facecolor='g', alpha=0.6)
```

```python
plt.title("Sample Visualization")
```

```python
#put together
plt.plot(x, y, '-')
plt.fill_between(x, y, 195, where=(y > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization")
plt.show()
```
